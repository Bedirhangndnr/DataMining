```python

def date_set(date):
    splitted_date=date.split()
    mouth_number={
        'OCAK':"01",
        'ŞUBAT':"02",
        'MART':"03",
        'NİSAN':"04",
        'MAYIS':"05",
        'HAZİRAN':"06",
        'TEMMUZ':"07",
        'AĞUSTOS':"08",
        'EYLÜL':"09",
        'EKİM':"10",
        'KASIM':"11",
        'ARALIK':"12"
    }
    
    for mouth_name, mouth_no in mouth_number.items():
        if(splitted_date[1]==mouth_name):
            splitted_date[1]=mouth_no
            to_convert_datatime=splitted_date[2]+"-"+splitted_date[1]+"-"+splitted_date[0]
            return pd.to_datetime(to_convert_datatime)     

covid_filled["date"]=covid_filled["date"].apply(lambda x: date_set(x))
#covid_filled=covid_filled[covid_filled.isnull().sum(axis=1)<3]
covid_filled=covid_filled[["date","daily_case"]]
covid_filled.info()
```

```python
g_mob.date=pd.to_datetime(g_mob.date)
merged_data=pd.merge(g_mob, 
         covid_filled, on="date")
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(20,15))
plt.plot(merged_data.date, merged_data.restaurants_cafes, "r--",
         merged_data.date, merged_data.daily_case,"b.")
```
